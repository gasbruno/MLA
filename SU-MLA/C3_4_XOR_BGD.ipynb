## 2-layers MLP for XOR learning using  Batch Gradient Descent

Reprendre le précédent modèle `XORModel` et réaliser un apprentissage en utilisant lé méthode de descente de gradient totale:
Les modification des poids sont calculées à chaque passage des exemples et mises de côté.
La modification des poids n'est opérée que lorsque tous les exemple du batch ont été présentés.
Le modèle doit donc cumuler les gradients dans une variable dédiée. 



Ecrire la classe `XORModel` permettant de réaliser un apprentissage comme suit:

```python
    xor_model = XORModel()
    lr = 0.001
    L = []
    num_epochs = 1000
    for epoch in range(num_epochs):
        xor_model.zero_grad()       # initialisation des accumulateurs du gradient
        loss = 0
        for ex in range(len(X)):
            x1, x2 = X[ex]
            y1, y2 = Y[ex]
            [ypred1, ypred2] = xor_model.forward(x1, x2, y1, y2)
            loss += xor_model.loss()
            xor_model.backward()

        L.append(loss / len(X))      # Normalisation / nombre d'exemples  
        xor_model.optimizerBGD(lr)
        print(f"Epoch {epoch} - Loss: {loss}")
```

## Solution

````python
import op

class XORModel:

    def __init__(self):
        self.L = 0

        # Desired outputs
        self.y2_1 = op.PlaceHolder(0)
        self.y2_2 = op.PlaceHolder(0)

        # Inputs
        self.x1 = op.PlaceHolder(0)
        self.x2 = op.PlaceHolder(0)

        # Layer 1 - Neuron 1
        self.w1_11 = op.Parameter(-1)
        self.w1_12 = op.Parameter(2)
        self.mul1_11 = op.Mul()
        self.mul1_12 = op.Mul()
        self.add1_11 = op.Add()
        self.b1_1 = op.Parameter(4)
        self.add1_12 = op.Add()
        self.sigma1_1 = op.Tanh()

        # Layer 1 - Neuron 2
        self.w1_21 = op.Parameter(-1)
        self.w1_22 = op.Parameter(1)
        self.mul1_21 = op.Mul()
        self.mul1_22 = op.Mul()
        self.add1_21 = op.Add()
        self.b1_2 = op.Parameter(-1)
        self.add1_22 = op.Add()
        self.sigma1_2 = op.Tanh()

        # Layer 2 - Neuron 1
        self.w2_11 = op.Parameter(-1)
        self.w2_12 = op.Parameter(2)
        self.mul2_11 = op.Mul()
        self.mul2_12 = op.Mul()
        self.add2_11 = op.Add()
        self.b2_1 = op.Parameter(4)
        self.add2_12 = op.Add()
        self.sigma2_1 = op.Tanh()

        # Layer 2 - Neuron 2
        self.w2_21 = op.Parameter(-1)
        self.w2_22 = op.Parameter(1)
        self.mul2_21 = op.Mul()
        self.mul2_22 = op.Mul()
        self.add2_21 = op.Add()
        self.b2_2 = op.Parameter(-1)
        self.add2_22 = op.Add()
        self.sigma2_2 = op.Tanh()

        # Error / Loss
        self.err1 = op.Sub()
        self.err2 = op.Sub()
        self.mse1 = op.Square()
        self.mse2 = op.Square()
        self.mseT = op.Add()

        # Weights accumulators
        self.w1_11_acc = 0
        self.w1_12_acc = 0
        self.b1_1_acc = 0
        self.w1_21_acc = 0
        self.w1_22_acc = 0
        self.b1_2_acc = 0
        self.w2_11_acc = 0
        self.w2_12_acc = 0
        self.b2_1_acc = 0
        self.w2_21_acc = 0
        self.w2_22_acc = 0
        self.b2_2_acc = 0


    def forward(self, x1, x2, y1, y2):
        # Desired outputs
        self.y2_1.set(y1)
        self.y2_2.set(y2)

        # Set inputs values
        self.x1.set(x1)
        self.x2.set(x2)

        # Forward Layer 1 neuron 1
        e1_11 = self.mul1_11(self.x1(), self.w1_11())
        e1_12 = self.mul1_12(self.x2(), self.w1_12())
        v1_10 = self.add1_11(e1_11, e1_12)
        v1_1 = self.add1_12(v1_10, self.b1_1())
        y1_1 = self.sigma1_1(v1_1)

        # Forward Layer 1 neuron 2
        e1_21 = self.mul1_21(self.x1(), self.w1_21())
        e1_22 = self.mul1_22(self.x2(), self.w1_22())
        v1_20 = self.add1_21(e1_21, e1_22)
        v1_2 = self.add1_22(v1_20, self.b1_2())
        y1_2 = self.sigma1_2(v1_2)

        # Forward Layer 2 neuron 1
        e2_11 = self.mul2_11(y1_1, self.w2_11())
        e2_12 = self.mul2_12(y1_2, self.w2_12())
        v2_10 = self.add2_11(e2_11, e2_12)
        v2_1 = self.add2_12(v2_10, self.b2_1())
        ypred_1 = self.sigma2_1(v2_1)

        # Forward Layer 2 neuron 2
        e2_21 = self.mul2_21(y1_1, self.w2_21())
        e2_22 = self.mul2_22(y1_2, self.w2_22())
        v2_20 = self.add2_21(e2_21, e2_22)
        v2_2 = self.add2_22(v2_20, self.b2_2())
        ypred_2 = self.sigma2_2(v2_2)

        # Error
        error1 = self.err1(ypred_1, self.y2_1())
        error2 = self.err2(ypred_2, self.y2_2())
        mserror1 = self.mse1(error1)
        mserror2 = self.mse2(error2)

        # Store loss
        self.L = self.mseT(mserror1, mserror2)

        # Return prediction
        return [ypred_1, ypred_2]

    def loss(self):
        return self.L
    
    def backward(self):
        g_L = 1

        # Backward Layer 2 neuron 1
        g_mserror1 = self.mseT.grad1(g_L)
        g_error1 = self.mse1.grad(g_mserror1)
        g_ypred_1 = self.err1.grad1(g_error1)
        g_v2_1 = self.sigma2_1.grad(g_ypred_1)
        g_v2_10 = self.add2_12.grad1(g_v2_1)
        self.g_b2_1 = self.add2_12.grad2(g_v2_1)
        g_e2_11 = self.mul2_11.grad1(g_v2_10)
        self.g_w2_11 = self.mul2_11.grad2(g_v2_10)
        g_e2_12 = self.mul2_12.grad1(g_v2_10)
        self.g_w2_12 = self.mul2_12.grad2(g_v2_10)

        # Backward Layer 2 neuron 2
        g_mserror2 = self.mseT.grad2(g_L)
        g_error2 = self.mse2.grad(g_mserror2)
        g_ypred_2 = self.err2.grad1(g_error2)
        g_v2_2 = self.sigma2_2.grad(g_ypred_2)
        g_v2_20 = self.add2_22.grad1(g_v2_2)
        self.g_b2_2 = self.add2_22.grad2(g_v2_2)
        g_e2_21 = self.mul2_21.grad1(g_v2_20)
        self.g_w2_21 = self.mul2_21.grad2(g_v2_20)
        g_e2_22 = self.mul2_22.grad1(g_v2_20)
        self.g_w2_22 = self.mul2_22.grad2(g_v2_20)

        # Backward Layer 1 neuron 1
        g_v1_1 = self.sigma1_1.grad(g_e2_11 + g_e2_21)
        g_v1_10 = self.add1_12.grad1(g_v1_1)
        self.g_b1_1 = self.add1_12.grad2(g_v1_1)
        self.g_w1_11 = self.mul1_11.grad2(g_v1_10)
        self.g_w1_12 = self.mul1_12.grad2(g_v1_10)

        # Backward Layer 1 neuron 2
        g_v1_2 = self.sigma1_2.grad(g_e2_12 + g_e2_22)
        g_v1_20 = self.add1_22.grad1(g_v1_2)
        self.g_b1_2 = self.add1_22.grad2(g_v1_2)
        self.g_w1_21 = self.mul1_21.grad2(g_v1_20)
        self.g_w1_22 = self.mul1_22.grad2(g_v1_20)

        # Accumulate gradients
        self.w2_11_acc += self.g_w2_11
        self.w2_12_acc += self.g_w2_12
        self.b2_1_acc += self.g_b2_1
        self.w2_21_acc += self.g_w2_21
        self.w2_22_acc += self.g_w2_22
        self.b2_2_acc += self.g_b2_2
        self.w1_11_acc += self.g_w1_11
        self.w1_12_acc += self.g_w1_12
        self.b1_1_acc += self.g_b1_1
        self.w1_21_acc += self.g_w1_21
        self.w1_22_acc += self.g_w1_22
        self.b1_2_acc += self.g_b1_2

    def zero_grad(self):
        """Reset weights accumulators"""
        self.w2_11_acc = 0
        self.w2_12_acc = 0
        self.b2_1_acc = 0
        self.w2_21_acc = 0
        self.w2_22_acc = 0
        self.b2_2_acc = 0
        self.w1_11_acc = 0
        self.w1_12_acc = 0
        self.b1_1_acc = 0
        self.w1_21_acc = 0
        self.w1_22_acc = 0
        self.b1_2_acc = 0

    def optimizerBGD(self, lr):
        """Batch Gradient Descent."""
        self.w2_11.update(self.w2_11() - lr * self.w2_11_acc)
        self.w2_12.update(self.w2_12() - lr * self.w2_12_acc)
        self.b2_1.update(self.b2_1() - lr * self.b2_1_acc)
        self.w2_21.update(self.w2_21() - lr * self.w2_21_acc)
        self.w2_22.update(self.w2_22() - lr * self.w2_22_acc)
        self.b2_2.update(self.b2_2() - lr * self.b2_2_acc)
        self.w1_11.update(self.w1_11() - lr * self.w1_11_acc)
        self.w1_12.update(self.w1_12() - lr * self.w1_12_acc)
        self.b1_1.update(self.b1_1() - lr * self.b1_1_acc)
        self.w1_21.update(self.w1_21() - lr * self.w1_21_acc)
        self.w1_22.update(self.w1_22() - lr * self.w1_22_acc)
        self.b1_2.update(self.b1_2() - lr * self.b1_2_acc)
````

## Go !

````python
# Définition des données d'apprentissage
X = [[1, 1], [1, -1], [-1, 1], [-1, -1]]
Y = [[0, 1], [1, 0], [1, 0], [0, 1]]

xor_model = XORModel()
lr = 0.001
L = []
num_epochs = 1000
for epoch in range(num_epochs):
    xor_model.zero_grad()       # initialisation des accumulateurs du gradient
    loss = 0
    for ex in range(len(X)):
        x1, x2 = X[ex]
        y1, y2 = Y[ex]
        [ypred1, ypred2] = xor_model.forward(x1, x2, y1, y2)
        loss += xor_model.loss()
        xor_model.backward()

    L.append(loss / len(X))      # Normalisation / nombre d'exemples  
    xor_model.optimizerBGD(lr)
    print(f"Epoch {epoch} - Loss: {loss}")
````

## Display the Loss

````python
import matplotlib.pyplot as plt
plt.plot(L)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('Training Loss over Iterations')
plt.show()
````

## Frontiers

````python
import numpy as np
import matplotlib.pyplot as plt

# Visualisation des résultats
def plot_decision_boundary(model, X, Y):
    X = np.array(X)
    Y = np.array(Y)
    # cycle over 2D grid
    plt.figure(figsize=(5, 5))
    for i in np.arange(-1.5, 1.5, 0.07):
        for j in np.arange(-1.5, 1.5, 0.07):
            # eval model on each grid point
            [ypred1, ypred2] = model.forward(i, j, 0, 0)
            if ypred1 > ypred2:
                plt.plot(i, j, ".r")
            else:
                plt.plot(i, j, ".b")

    for i in range(Y.shape[0]):
        if Y[i][0] == 1 and Y[i][1] == 0:
            plt.plot(X[i, 0], X[i, 1], "ro")
        else:
            plt.plot(X[i, 0], X[i, 1], "bo")
    plt.show()

plot_decision_boundary(xor_model, X, Y)
````
